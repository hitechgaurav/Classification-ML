```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

```python
data = pd.read_csv("E:\Python All in One\DATA_SETS\Machine Learning A-Z Template Folder\Part 3 - Classification\Section 15 - K-Nearest Neighbors (K-NN)\K_Nearest_Neighbors\Social_Network_Ads.csv")
```

```python
data.head()
```

```python
#Separating the dependent and independent variables
x = data.iloc[:, 2:4].values
y = data.iloc[:,4].values
```

```python
#Applying Feature Scaling in x parameter
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x = sc_x.fit_transform(x)
x
```

```python
#Splitting training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .25, random_state = 0)
```

```python
for i in x_train:
    print(i[0])
```

```python
#Lets Visualize training plots
pt.figure(figsize = (10,5))
for i in x_train:
    pt.scatter(i[0], i[1], c="indigo")
pt.xlabel("Age")
pt.ylabel("Salary")
pt.title("Age vs Salary")
pt.show()
```

```python
#Building the k-nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)

classifier.fit(x_train, y_train)
```

```python
#Predicting the Values
y_pred = classifier.predict(x_test)
```

```python
#Evaluating our model
score = classifier.score(x_test, y_test)*100

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print(score)

print(cm)
```
